```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
num_particles = [1000, 5000, 10000, 50000, 100000, 500000]

times_serial = [0.0409785, 0.198229, 0.685555, 3.39095,  6.97389, 34.8343]

times_parallel = [0.0516941, 0.199959,  0.389617, 1.83601, 3.64478, 17.7592]
```

```python
plt.plot(np.log10(num_particles), np.log10(times_serial), '-o', label='serial', color = 'red')
plt.plot(np.log10(num_particles), np.log10(times_parallel), '-o', label='openMP')

plt.legend()
plt.xlabel("Log Number of particles")
plt.ylabel("Log Time [seconds]")
plt.title("Computation time vs Number of particles")
plt.savefig("Plot1.png")
```

```python
## Strong scaling OpenMP

# 1000, 8000, 24000, 100000

num_threads = [2,4,8,16,32,64]
times_1000 = [0.0474597, 0.0550374, 0.0464975, 0.045065, 0.0477323, 0.052104]
times_8000 = [0.498041, 0.495133, 0.365628, 0.318621, 0.312884, 0.30343]
times_16000 = [1.05448, 1.01898, 0.75527, 0.644763, 0.616155, 0.593444]
times_32000 = [2.08905, 2.04583, 1.48433, 1.22643, 1.19766, 1.19688]
times_64000 = [4.19711, 3.85332, 3.09562, 2.49114, 2.39055, 2.35936]
times_128000 = [7.95375, 7.54307, 5.87645, 4.90519, 4.82694, 4.66433]


plt.plot(num_threads, times_1000, '-o', label='1000 particles')
plt.plot(num_threads, times_8000, '-o', label='8000 particles')
plt.plot(num_threads, times_16000, '-o', label='16000 particles')
plt.plot(num_threads, times_32000, '-o', label='32000 particles')
plt.plot(num_threads, times_64000, '-o', label='64000 particles')
plt.plot(num_threads, times_128000, '-o', label='128000 particles')

#plt.plot(np.log10(num_particles), np.log10(times_parallel), label='openMP')

plt.legend()
plt.xlabel("Number of Threads")
plt.ylabel("Time [seconds]")
plt.title("Computation time vs Number of Threads (Strong Scaling)")
plt.savefig("Plot2_strong_scaling_noLOG.png")
```

```python
## Weak scaling OpenMP

# 1000 particles per thread
# 2000 particles per thread

num_threads = [2,4,8,16,32,64]
num_particles_1 = [2000, 4000, 8000, 16000, 32000, 64000]
num_particles_2 = [4000, 8000, 16000, 32000, 64000, 128000]

times_1000_w = [0.0883038, 0.24119, 0.372667, 0.61988, 1.19834, 2.36663]
times_2000_w = [0.254549, 0.49602, 0.745622, 1.2172, 2.37195, 4.61955]


plt.plot((num_threads), (times_1000_w), '-o', label='1000 particles per thread')
plt.plot((num_threads), (times_2000_w), '-o', label='2000 particles per thread')


#plt.plot(np.log10(num_particles), np.log10(times_parallel), label='openMP')

plt.legend()
plt.xlabel("Number of Threads")
plt.ylabel("Time [seconds]")
plt.title("Computation time vs Number of Threads (Weak Scaling)")
plt.savefig("Plot3_weak_scaling_noLOG.png")
```
